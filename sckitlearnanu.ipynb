```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
x,y = data.data, data.target
print("fitur : ", data.feature_names)
print("target : ", data.target_names)
print("shape x : ", x.shape, "shape y  : ", y.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print("Data Train:", x_train.shape, "Data Test:", x_test.shape)
```

```python
# normalisasi data (scalling)
scaler = StandardScaler() #inisialisasi scalar
X_train_scaled = scaler.fit_transform(x_train) 
X_test_scaled = scaler.transform(x_test)
print("sebelum scaling  : ", x_train[:5])
print("sesudah scaling  : ", X_train_scaled)
```

```python
encoder = LabelEncoder()
y_train_encoded = encoder.fit_transform(y_train)
y_test_encoded = encoder.transform(y_test)
model = DecisionTreeClassifier()
model.fit(X_train_scaled,y_train)
y_pred = model.predict(X_test_scaled)
print("pediksi :    ", y_pred)
print("label asli : ", y_test)
```

```python

accuracy = accuracy_score(y_test, y_pred)
print(f"Akurasi Model: {accuracy:.2f}")
```
